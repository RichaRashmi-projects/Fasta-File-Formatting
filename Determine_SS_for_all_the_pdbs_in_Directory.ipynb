```python
import glob
```

```python

infiles = glob.glob("/content/drive/MyDrive/sde2_collaboration/rosetta_files/pdb_files/*.pdb")
infiles
```

```python

outfiles = [sub.replace(".pdb", "_dssp.txt") for sub in infiles]
outfiles
```

```python
in_out_dict = dict(zip(infiles, outfiles))

in_out_dict.items()
```

```python
import os
 #os.system("rm /content/drive/MyDrive/sde2_collaboration/rosetta_files/pdb_files/*.txt")
```

```python
for infile , outfile in in_out_dict.items():
  #command1 = "rm /content/drive/MyDrive/sde2_collaboration/rosetta_files/pdb_files/*.txt"
  command = "dssp -i "+infile+ " -o "+ outfile
  os.system(command)
  print(command)
```
